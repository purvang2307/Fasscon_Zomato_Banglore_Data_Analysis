## Introduction to Zomato Bangalore Restaurant Data Analysis

### Background:

[Zomato](https://www.zomato.com/) is a popular online platform that provides information and reviews about restaurants, cafes, and food establishments in various cities around the world. With its vast database of restaurant listings, Zomato serves as a valuable resource for both food enthusiasts and businesses alike.

### About the Dataset:

The dataset used in this analysis focuses on restaurants in Bangalore, India, collected from Zomato. Bangalore, often referred to as the "Silicon Valley of India," is known for its vibrant food culture and diverse culinary scene. With thousands of restaurants catering to a wide range of tastes and preferences, Bangalore provides an ideal setting for exploring trends and patterns in the food industry.

### Objective:

The objective of this analysis is to delve into the Zomato Bangalore restaurant data and extract meaningful insights regarding various aspects such as cuisine types, restaurant ratings, cost factors, and more. By leveraging data science techniques, we aim to uncover trends, preferences, and patterns that can offer valuable insights for consumers, restaurant owners, and food industry stakeholders.

### Key Components:

- **Data Exploration:** We will begin by exploring the structure and contents of the dataset, gaining a deeper understanding of the variables and their significance.
  
- **Data Cleaning:** Next, we will preprocess the data to handle missing values, remove duplicates, and ensure consistency, laying the foundation for reliable analysis.
  
- **Exploratory Data Analysis (EDA):** Through visualizations and statistical analysis, we will investigate various aspects of the dataset, identifying trends, correlations, and outliers.
  
- **Insights and Interpretations:** Based on our analysis, we will draw insights regarding popular cuisines, factors influencing restaurant ratings, pricing trends, and other relevant observations.

### Why This Analysis Matters:

Understanding the dynamics of the restaurant industry in Bangalore can provide valuable insights for multiple stakeholders:

- **Consumers:** Make informed decisions about dining options based on factors such as cuisine preferences, pricing, and ratings.
  
- **Restaurant Owners:** Gain insights into customer preferences, competitive landscape, and factors influencing customer satisfaction to optimize their offerings.
  
- **Food Industry Analysts:** Explore trends and patterns in the restaurant sector to identify opportunities, challenges, and emerging market trends.

**Issue with cuisines :<br>**
* **The data string is so long and it contains multiple cuisines type.**

```python
df['cuisines'].value_counts().head(10)
```

```python
def cleaning_cuisines(val):
    if (val == 'North Indian, Chinese'):
        return 'MultiCusine'
    elif ((val == 'Bakery, Desserts') or (val== 'Desserts')):
        return 'Sweat Dish'
    elif ((val == 'South Indian') or (val == 'Biryani')):
        return 'South Indian'
    elif (val == 'Fast Food'):
        return 'Fast Food'
    elif (val == 'Cafe'):
        return 'Cafe'
    elif (val == 'North Indian'):
        return 'North Indian'
    else :
        return 'Others'
    
df['cuisines'].apply(cleaning_cuisines).value_counts()
```

```python
df['cuisines']=df['cuisines'].apply(cleaning_cuisines)
df.head()
```

```python
df['rate'].unique()
```

```python
def cleaning_rate(val):
    if (val == "NEW" or val == '-') :
        return np.nan
    else:
        return float(str(val).split('/')[0])

df['rate']=df['rate'].apply(cleaning_rate)
```

```python
df['rate'].median()
```

```python
def impute_nan(df,col):
    df[col + '_median']=df[col].fillna(df[col].median())
    df[col + '_random']=df[col]
    random_sample=df[col].dropna().sample(df[col].isnull().sum(),random_state=0)
    random_sample.index=df[df[col].isnull()].index
    df.loc[df[col].isnull(), col+'_random']=random_sample
```

```python
impute_nan(df,'rate')
```

```python
df.head()
```

```python
df['rate'].plot(kind='kde',color='yellow',label='rate')
df['rate_median'].plot(kind='kde',color='green',label='rate_median')
plt.legend()
plt.show()
```

```python
df['rate'].plot(kind='kde',color='yellow',label='rate')
df['rate_random'].plot(kind='kde',color='blue',label='rate_random')
plt.legend()
plt.show()
```

* **Here we can see that filling nan values for rate using either mode or random sample imputation gives the exact distribution.<br>**
* **So we can use either mode or random sample imputation.<br>**
* **We will use random sample imputation here.<br>**

```python
df['rate']=df['rate_random']
```

```python
df.isna().sum()/df.shape[0]
```

```python
df.drop(columns=['rate_median','rate_random'],axis=1,inplace=True)
```

```python
df['cost_for_2'].plot(kind='kde',color='yellow',label='cost_for_2')
df['cost_for_2_random'].plot(kind='kde',color='blue',label='cost_for_2_random')
plt.legend()
plt.show()
```

* **To many categories are there.<br>**

```python
rest_type_less=rest_type[rest_type.counts < 650]
rest_type_less
```

```python
def cleaning_restaurant_type(val):
    if (val=='Takeaway, Delivery'):
        return 'Delivery'
    elif (val=='Casual Dining, Bar'):
        return 'Bar'
    else:
        return val
    
df['rest_type']=df['rest_type'].apply(cleaning_restaurant_type)
```

## Data Visualization

```python
plt.figure(figsize=(10,6))
sns.countplot(x= df['rest_type'])
plt.xlabel(' Restaurant Type')
plt.ylabel('Count')
plt.title('Countplot for rest_type')
plt.show()
```

```python
counts = df.groupby(['location', 'rest_type']).size().reset_index(name='count')
max_counts = counts.groupby('location').agg({'count': 'max'}).reset_index()
max_counts = pd.merge(max_counts, counts, on=['location', 'count'], how='left')
max_counts=max_counts[['location', 'rest_type', 'count']].sort_values(by='count',ascending=False).reset_index(drop=True)
max_counts
```

```python
plt.figure(figsize=(14,9))
sns.barplot(x=max_counts['location'],y= max_counts['count'],orient='v')
plt.xticks(rotation=90)
plt.xlabel('Location')
plt.ylabel('count')
plt.title('Bar plot for Location Wise Maximum Count for Restaurant Type')
for i in range(len(max_counts)):
    plt.text(i, max_counts['count'][i] + 0.5, max_counts['rest_type'][i], ha='center',fontsize=6)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x= df['type'])
plt.xlabel('Type')
plt.ylabel('Count')
plt.title('Countplot for Type')
plt.show()
```

```python
type_counts = df.groupby(['location', 'type']).size().reset_index(name='count')
max_type_counts = type_counts.groupby('location').agg({'count': 'max'}).reset_index()
max_type_counts = pd.merge(max_type_counts, type_counts, on=['location', 'count'], how='left')
max_type_counts=max_type_counts[['location', 'type', 'count']].sort_values(by='count',ascending=False).reset_index(drop=True)
max_type_counts
```

```python
plt.figure(figsize=(14,9))
sns.barplot(x=max_type_counts['location'],y= max_type_counts['count'],orient='v')
plt.xticks(rotation=90)
plt.xlabel('Location')
plt.ylabel('count')
plt.title('Bar plot for Location Wise Maximum Count for Type')
for i in range(len(max_type_counts)):
    plt.text(i, max_type_counts['count'][i] + 0.5, max_type_counts['type'][i], ha='center',fontsize=8)
plt.show()
```

```python
cost_avg_df=df.groupby('location')['cost_for_2'].median().round(2).reset_index()
cost_avg_df.columns=['location','Avg_Cost']
cost_avg_df=cost_avg_df.sort_values(by="Avg_Cost",ascending=False).reset_index(drop=True)
cost_avg_df
```

```python
plt.figure(figsize=(14,9))
sns.barplot(x=cost_avg_df['location'],y= cost_avg_df['Avg_Cost'],orient='v')
plt.xticks(rotation=90)
plt.xlabel('Location')
plt.ylabel('Average Cost')
plt.title('Average Cost for 2 People Location Wise')
for i in range(len(cost_avg_df)):
    plt.text(i , cost_avg_df['Avg_Cost'][i]+0.5 , cost_avg_df['Avg_Cost'][i], ha='center',fontsize=8)
plt.show() 
```

```python
cost_avg_rest_df=df.groupby('rest_type')['cost_for_2'].median().round(2).reset_index()
cost_avg_rest_df.columns=['rest_type','Avg_Cost']
cost_avg_rest_df=cost_avg_rest_df.sort_values(by="Avg_Cost",ascending=False).reset_index(drop=True)
cost_avg_rest_df
```

```python
plt.figure(figsize=(14,9))
sns.barplot(x=cost_avg_rest_df['rest_type'],y= cost_avg_rest_df['Avg_Cost'],orient='v')
plt.xticks(rotation=90)
plt.xlabel('Restaurant Type')
plt.ylabel('Average Cost')
plt.title('Average Cost for 2 People Restaurant Type Wise')
for i in range(len(cost_avg_rest_df)):
    plt.text(i , cost_avg_rest_df['Avg_Cost'][i]+0.5 , cost_avg_rest_df['Avg_Cost'][i], ha='center',fontsize=8)
plt.show() 
```
